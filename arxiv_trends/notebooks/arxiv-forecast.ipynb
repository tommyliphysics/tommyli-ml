## Polynomial regression

Clearly our data has a trend; I will be interested in long-term forecasting. Autoregression methods are not suitable for this task since the time series is not stationary. The simplest approach to modelling the trend is polynomial regression, however we must be careful about using this method for extrapolation, as the highest power will dominate with increasing time. We can use the order of the polynomial as a hyperparameter and tune it using a validation set consisting of the last 24 months of the training data. Let's get the validation mean squared errors for different orders:

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline

def score_predictions(degree, X_train_, X_test_, y_train_, y_test_):
    pipe = Pipeline([('polynomial_features', PolynomialFeatures(degree=degree)), ('linear_fit', LinearRegression())])
    pipe.fit(X_train_, y_train_)
    y_pred_train = pipe.predict(X_train_)
    y_pred_test = pipe.predict(X_test_)

    return y_pred_train, y_pred_test, mean_squared_error(y_test_, y_pred_test)
```

```python
val = train.iloc[-24:]
train_ = train.iloc[:-24]

X_train = train_[['months_since_Jan_2000']]
y_train = train_['machine learning']

X_val = val[['months_since_Jan_2000']]
y_val = val['machine learning']

y_pred_train, y_pred_val, mse = score_predictions(5, X_train, X_val, y_train, y_val)
```

```python
y_pred_trains = []
y_pred_vals = []
errors = []

for degree in range(50):
    y_pred_train, y_pred_val, error = score_predictions(degree, X_train, X_val, y_train, y_val)
    y_pred_trains.append(y_pred_train)
    y_pred_vals.append(y_pred_val)
    errors.append({'degree': degree, 'error': np.sqrt(error)})
```

```python
fig,ax = plt.subplots(figsize=(10,6))
ax.set_yscale('log')
sns.lineplot(x=np.arange(len(errors)), y=[errors[deg]['error'] for deg in range(len(errors))])
```

This shows the danger in using polynomial regression to perform extrapolation from long-term trends.

## Deep learning

We will next look at using neural networks to perform long-term forecasting. Our strategy will be the following:
* Break up a long time series into smaller periods each consisting of an "X" (feature) segment and a "y" (target) segment.
* The model will try to predict the target given the feature.
* We will use all 500 keywords for training data, so the model learns the time evolution of each keyword but is agnostic as to its meaning.
We will compare pure feed-forward networks with LSTMs and vary the number of layers and neurons, and different activations (linear, relu, tanh).

First, we have to create a function that generates subsamples from a single time series:

```python
def create_samples(time_series, steps_X, steps_y, overlap):
    start = 0
    samples_X = []
    samples_y = []
    while start + steps_X + steps_y <= len(time_series):
        samples_X.append(time_series[start:start+steps_X])
        samples_y.append(time_series[start+steps_X:start+steps_X+steps_y])
        start += steps_X + steps_y - overlap
    return np.array(samples_X), np.array(samples_y)
```

We will choose a period of 96 months for the feature and 24 months for the target, and allow each 120 month period to overlap with the previous one by 12 months.

```python
steps_X = 24*4
steps_y = 24
overlap = steps_X+steps_y-12
```

```python
series = np.arange(3,267)
samples_X, samples_y= create_samples(series, steps_X, steps_y, overlap)
print('samples_X:', samples_X)
print('samples_y:', samples_y)
```

```python
from sklearn.preprocessing import MinMaxScaler

def prepare_input(data, scaler=None):
    X = []
    y = []

    for col in data.columns:
        samples_X, samples_y = create_samples(data[col].to_numpy(), steps_X, steps_y, overlap)
        if(len(X) == 0):
            X = samples_X
            y = samples_y
        else:
            X = np.concatenate((X, samples_X), axis=0)
            y = np.concatenate((y, samples_y), axis=0)
    if scaler == None:
        scaler = MinMaxScaler()
        X_scaled = scaler.fit_transform(X)
        X_scaled = np.reshape(X_scaled, (X_scaled.shape[0], X_scaled.shape[1], 1))

        return X_scaled, y, scaler
    else:
        X_scaled = scaler.transform(X)
        X_scaled = np.reshape(X_scaled, (X_scaled.shape[0], X_scaled.shape[1], 1))

        return X_scaled, y
```

```python
def plot_training(model_names, histories):
    fig, axes = plt.subplots(len(model_names),len(model_names[0]),figsize=(10,8))

    for n in range(len(model_names)):
        for m in range(len(model_names[0])):
            history = histories[n][m]
            sns.lineplot(x=np.arange(len(history.history['mse'])), y=history.history['mse'], ax=axes[n,m], label=f"mse (train), final={history.history['mse'][-1]:.3e}")
            sns.lineplot(x=np.arange(len(history.history['val_mse'])), y=history.history['val_mse'], ax=axes[n,m], label =f"mse (val), final={history.history['val_mse'][-1]:.3e}")
            axes[n,m].set_yscale('log')
            axes[n,m].set_title(model_names[n][m])
```

### Feed-forward networks

```python
from keras.models import Model
from keras.layers import Input, Dense

def build_FF(n_layers, units, activation_hidden='linear', activation_final='linear'):
    input = Input(shape=(steps_X,))
    x = Dense(units, activation=activation_hidden)(input)
    for n in range(n_layers-1):
        x = Dense(units, activation=activation_hidden)(x)
    output = Dense(steps_y, activation=activation_final)(x)
    model = Model(inputs=input, outputs=output)
    model.compile(optimizer='adam', loss='mse', metrics=['mse'])
    return model

def train_model(model):
    return model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data = (X_val,y_val),verbose=False)
```

```python
histories = {}
models = {}

epochs=100
batch_size=32
```

```python
models['FF_1_4'] = build_FF(1,4)
models['FF_1_16'] = build_FF(1,16)
models['FF_4_4'] = build_FF(4,4)
models['FF_4_16'] = build_FF(4,16)
```

```python
from tqdm import tqdm
for model_name in tqdm(models):
    histories[model_name] = train_model(models[model_name])
```

Let's add dropout:

```python
from keras.layers import Dropout

def build_FF_with_dropout(n_layers, units, dropout, activation_hidden='linear', activation_final='linear'):
    input = Input(shape=(steps_X,))
    x = Dense(units, activation=activation_hidden)(input)
    for n in range(n_layers-1):
        x = Dense(units, activation=activation_hidden)(x)
    x = Dropout(dropout)(x)
    output = Dense(steps_y, activation=activation_final)(x)
    model = Model(inputs=input, outputs=output)
    model.compile(optimizer='adam', loss='mse', metrics=['mse'])
    return model
```

```python
models['FF_1_4'] = build_FF_with_dropout(1,4, 0.5)
models['FF_1_16'] = build_FF_with_dropout(1,16, 0.5)
models['FF_4_4'] = build_FF_with_dropout(4,4, 0.5)
models['FF_4_16'] = build_FF_with_dropout(4,16, 0.5)
```

```python
from tqdm import tqdm
for model_name in tqdm(models):
    histories[model_name] = train_model(models[model_name])
```
